使用了堆栈的数据结构存放遍历的表项下标和表项数据，避免使用递归。

下面是对应的Python实现：

````python
import numpy as np


def construct_optional_BST(root):
    """
    construct BST using table root.
    :root  table root
    """
    shape = root.shape  # shape of the table root.
    s1 = []
    s2 = []             # consider a list as a stack.
    num = 0
    first = True
    s1.append((1, shape[1]))  # same function of push()
    index = s1.pop()
    while index:
        if index[0] <= index[1]:
            r = root[index[0] - 1, index[1] - 1]
        else:
            r = index[1]

        if first:
            first = False
            s2.append(r)
            print('k%d为根' % (r, ))
            s1.append((r + 1, shape[1]))
            s1.append((1, r - 1))
        else:
            if index[0] <= index[1]:
                num = s2.pop()
                if r < num:
                    print('k%d是k%d的左孩子' % (r, num))
                    s2.append(num)
                else:
                    print('k%d是k%d的右孩子' % (r, num))
                s2.append(r)
                s1.append((r + 1, index[1]))
                s1.append((index[0], r - 1))
            else:
                num = s2.pop()
                if r < num:
                    print('d%d是k%d的左孩子' % (index[1], num))
                    s2.append(num)
                else:
                    print('d%d是k%d的右孩子' % (index[1], num))
        try:
            index = s1.pop()
        except:
            index = None


if __name__ == '__main__':
    root = np.array([[1, 1, 2, 2, 2], 
                     [0, 2, 2, 2, 4], 
                     [0, 0, 3, 4, 5],
                     [0, 0, 0, 4, 5], 
                     [0, 0, 0, 0, 5]], 
                     'i')

    construct_optional_BST(root)
````
