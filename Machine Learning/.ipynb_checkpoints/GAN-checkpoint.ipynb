# 1  介绍
深度学习的前景在于发现一个丰富、分层的模型，这个模型可以用来表示所有类型的数据分布。目前机器学习比较成功的模型都是一类————判别模型，即将一个高维的输入映射到一个标签类别上。这些模型主要建立在反向传播算法和dropout算法的基础上。

但是深度生成模型中常会遇到一些难以计算的概率和复杂的决策问题，所以常常要用到×××的方法。而且因为分段线性单元在生成模型中很难被利用。本文提出了一种方法来客服上述的困难。

与其余的生成模型不同的是，对抗网络框架中的生成模型与一个判别模型进行对抗。这个对抗模型的目的就是使尽最大可能区分一个输入是来自真实样本还是生成的样本。在文中这被形象地比喻为是警察和制造假币的罪犯之间的对抗，警察尽可能分辨出纸币的真假，而罪犯尽可能印出更以假乱真的假币。当警察（判别模型）的分辨出假币时，刺激罪犯（生成模型）提高假币的印刷技术；当警察将假币误认为假币为真时，就刺激警察提高分辨能力。这种对抗机制使得双方能力得到增强，最终使得制造出的假币真假难辨，警察认为其真假概率都为50%

生成对抗模型的好处是不需要近似接口和马尔可夫链

# 2  相关工作
一些带潜变量非直接的图形模型，如受限玻尔兹曼机（RBNs）和深度玻尔兹曼机（DBMs）和它们的一些变体。这些方法都可以被看做是许多潜在函数的乘积，再通过对所有随机变量求和或求积分来归一化。这类方法难以定量计算，虽然可以用马尔可夫链蒙特卡洛方法来估计，但混合使得基于MCMC的算法产生了很多问题。

深度置信网络（DBNs）是一个包含有一个非直接层和多个直接层的混合模型。虽然存在快速近似的分层训练标准，但DBNs会产生与无向和有向模型相关的计算困难。

另外有一些方法，比如score matching和噪声对比估计（NCE）。这两种方法都要求将学习概率密度分析指定为归一化常数，但对于一些含有多个潜变量层的生成模型来说，这是一件不可能的事情。去噪自动编码器和收缩自动编码器的学习规则与score matching应用于受限玻尔兹曼机相似。在NCE中，采用歧视性训练标准来拟合生成模型。 然而，生成模型本身不是用一个单独的判别模型来拟合，而是用来区分生成的数据与固定噪声分布的样本。因为NCE用的是一个固定的噪声分布，所以学习过程在模型的一部分可观察到的变量接近正确分布时就会显著减慢。

最后，一些技术不涉及显式定义概率分布，而是训练一个生成机器从所需分布中抽取样本。 这类方法的优点在于它们可以用反向传播的方法来设计和训练。最近的一些有成效的工作包括生成随机网络（GSN）框架，它继承和扩展了去噪声自动编码器，这两者都可以被看做是参数化的马尔可夫链。与GSN相比，GAN不需要用马尔可夫链采样，因为GAN在生成中不需要反馈回路，它能更好地利用分段线性单元。这可以提高反向传播的性能，但是在反馈环路中使用时会遇到无限激活的问题。
